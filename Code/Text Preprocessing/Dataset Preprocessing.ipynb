# Cleaning UNTM Dataset

```python
#load the dataset
df = pd.read_excel("/content/drive/MyDrive/UNTM Dataset.xlsx") # add your repository
```

```python
df.head()
```

```python
#total documents
print(len(df))
```

```python
#No of documents that has missing value
df[df['News'].isnull().values]
```

```python
#remove rows that has missing value
df.dropna(inplace=True)
```

```python

print(len(df))
```

```python
#Remove rows that has word frequency less than 5
rows_to_remove = df[df['News'].apply(lambda x: len(x.split()) < 5)].index
df.drop(rows_to_remove, inplace=True)

print(len(df))
```

```python
# to remove the text before the specified pattern
df['Clean_Data'] = df['News'].str.replace(r'^(?:.*?)(\(اُردو پوائنٹ اخبارتازہ ترین -)', r'\1', regex=True)
```

```python
df.head()
```

```python
#Remove irrelavant information in brackets
def remove_brackets_with_text(text):
    return re.sub(r"\([^()]*\)", "", text)
df['Clean_Data'] = df['Clean_Data'].apply(remove_brackets_with_text)
```

```python
df.head()
```

```python
# to remove URLs
def remove_urls(text):
    return re.sub(r'http\S+', '', text)
df['Clean_Data'] = df['Clean_Data'].apply(remove_urls)
```

```python
# Function to remove email addresses
def remove_emails(text):
    return re.sub(r'\S+@\S+', '', text)

df['Clean_Data'] = df['Clean_Data'].apply(remove_emails)
```

```python
#Remove Urdu Punctuation
def remove_urdu_punctuation(text):
    pattern = r'[؛؟،٫٬‘’“”«»!"٪&\'\(\)\*\+,-./:;<=>؟@\[\\\]^_`{|}~]'

    cleaned_text = re.sub(pattern, '', text)

    return cleaned_text
```

```python
df['Clean_Data'] = df['Clean_Data'].apply(remove_urdu_punctuation)
```

```python
df.Clean_Data = [doc.replace('\ufeff', '') for doc in df.Clean_Data]
```

```python
#remove digits
df.Clean_Data=df.Clean_Data.apply(lambda x: re.sub('\d','',x))
```

```python
#Remove English Alphabets
df.Clean_Data=df.Clean_Data.apply(lambda x: re.sub(r'[a-zA-Z]','',x))
```

```python
df.to_csv('Clean_UNTM.csv', index=False,  encoding='utf-8-sig')
```

# Cleaning UDC Dataset

```python
with open('/content/drive/MyDrive/BigCorpus.txt', 'r', encoding='utf-8') as file:
    file_contents = file.read()
#print(file_contents)
    documents=[]
    file=""

    for line in file_contents.split('\n'):
        file = file + " " + line
        if line.find('|') > -1:
            documents.append(file.replace('|',''))
            file = ""

print(len(documents))
```

```python
# Assign categories to each document
categories = ['Business', 'Entertainment', 'Sports', 'Health', 'Weird']
actual_categories = []

for i, doc in enumerate(documents):
    if i < 200:
        category = categories[0]
    elif i < 410:
        category = categories[1]
    elif i < 610:
        category = categories[2]
    elif i < 810:
        category = categories[3]
    else:
        category = categories[4]

    actual_categories.append(category)

df1 = pd.DataFrame({'Document': documents, 'Category': actual_categories})
print(df1)
```

```python
#Remove rows that has word frequency less than 4
rows_to_remove = df1[df1['Document'].apply(lambda x: len(x.split()) < 5)].index
df1.drop(rows_to_remove, inplace=True)
#remaining documents for experiments
print(len(df1))
```

```python
#remove emails
df1['Clean_Data'] = df1['Clean_Data'].apply(remove_emails)
```

```python
df1.Clean_Data = [doc.replace('\ufeff', '') for doc in df1.Clean_Data]
```

```python
#remove digits
df1.Clean_Data=df1.Clean_Data.apply(lambda x: re.sub('\d','',x))
```
