
# Importing required dependencies

Pandas will be used to create a Dataframe and handle the csv file. Numpy will be used for the faster computation of arrays to save time. re library will be used for the cleaning of data.

```python
import pandas as pd
import numpy as np
import re
```

##DataFrame




```python
#load clean Dataset
df = pd.read_csv('/content/drive/MyDrive/Clean_UNTM.csv', encoding='utf-8')
```

```python
def remove_nonbreaking_space(text):
    return re.sub(r'\xa0', ' ', text)

df['Clean_Data'] = df['Clean_Data'].apply(remove_nonbreaking_space)
```

```python
df.head()
```

```python
week_days = ["جمعرات", "اتوار", "بدھ", "ہفتہ", "جمعہ",  'پير', "منگل"]
monthly_names = ["دسمبر", "نومبر", "اکتوبر", "ستمبر", "اگست", "جولائی", "جون", "مئی", "اپريل", "مارچ", "فروری", "جنوری"]

df['Clean_Data'] = df['Clean_Data'].replace('|'.join(week_days), '', regex=True)
df['Clean_Data'] = df['Clean_Data'].replace('|'.join(monthly_names), '', regex=True)
```

```python
df['Clean_Data'] = df['Clean_Data'].str.replace('شیئرٹویٹشیئرای میلتبصرے مزید شیئر', '')
```

```python
for document in df['Clean_Data'].iloc[5:15]:
    print(document)
    print('---')
```

```python
documents = df['Clean_Data'].values.tolist()
print((documents[1]))
```

##Stopwords Removal

```python
#remove stopwords
import nltk
stopwords=pd.read_csv('/content/drive/MyDrive/stopwords.txt',names=['List'])
# stopwords['List']

stopwords_ur=[]
for li in stopwords['List']:
  stopwords_ur.append(li)
print(len(stopwords_ur))
print(stopwords_ur)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#countvectorizer with stopwords
vectorizer = CountVectorizer(stop_words= stopwords_ur)
X = vectorizer.fit_transform(documents)
```

##Seed words using features Selection Method
 we evaluate five feature selection techniques — Balance Accuracy Measure , GINI Index , Normalized Difference Measure, Information Gain , and Chi-Squared  — to derive seed sets using one-verses-all classicafication method


1.  Use Information Gain , and Chi-Squared from sklearn library
2.  Run Balance Accuracy Measure , GINI Index , Normalized Difference Measure from FS (Feature Selection) Python file.



```python
from sklearn.feature_selection import SelectKBest
from functools import partial
from feature_selection import FeatureSelection
```

```python
fs=FeatureSelection() # Function calling
```

```python

top_10_words_list = []

# Define the categories
categories = ['Sports', 'Weird', 'Business','Showbiz', 'Technology', 'Politics','Health']

for category in categories:

    df['Binary'] = df['Category'].apply(lambda x: 1 if x == category else 0)

    selector = SelectKBest(score_func=partial(fs.ndm), k=10)
    X_selected = selector.fit_transform(X, df['Binary'])

    feature_names = vectorizer.get_feature_names_out()[selector.get_support(indices=True)]

    top_10_words_list.append(feature_names.tolist())

print(top_10_words_list)
```

```python
from sklearn.feature_selection import SelectKBest, chi2

top_10_words_list = []

# Define the categories
categories = ['Sports', 'Weird', 'Business','Showbiz', 'Technology', 'Politics','Health']

for category in categories:
    df['Binary'] = df['Category'].apply(lambda x: 1 if x == category else 0)

    selector = SelectKBest(score_func=chi2, k=10)

    # Convert the feature values into non-negative integers
    X_non_negative = X - X.min()

    X_selected = selector.fit_transform(X_non_negative, df['Binary'])

    feature_names = vectorizer.get_feature_names_out()[selector.get_support(indices=True)]

    top_10_words_list.append(feature_names.tolist())

print(top_10_words_list)
```

```python
#top 10 seed words of each category
from sklearn.feature_selection import SelectKBest, mutual_info_classif

top_10_words_list = []

# Define the categories
categories = ['Sports', 'Weird', 'Business','Showbiz', 'Technology', 'Politics','Health']

for category in categories:

    df['Binary'] = df['Category'].apply(lambda x: 1 if x == category else 0)

    selector = SelectKBest(score_func=mutual_info_classif, k=10)
    X_selected = selector.fit_transform(X, df['Binary'])

    feature_names = vectorizer.get_feature_names_out()[selector.get_support(indices=True)]

    top_10_words_list.append(feature_names.tolist())

print(top_10_words_list)
```
