```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python

# selected_category = 'Business'
# category_df = df[df['Category'] == selected_category]

documents = df['News'].tolist()
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)
feature_names = vectorizer.get_feature_names_out()
total_tokens = X.sum()
unique_tokens = len(feature_names)

# Print the results
# print(f"Category: {selected_category}")
print(f"Total number of tokens (words): {total_tokens}")
print(f"Number of unique tokens (words): {unique_tokens}")
```

```python
#Average, min and max length
# business_documents = df[df['Category'] == 'Sports']['Document']
documents = df['News'].tolist()
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)
feature_names = vectorizer.get_feature_names_out()
document_lengths = X.sum(axis=1)

# Calculate minimum, maximum, and average document lengths
min_length = document_lengths.min()
max_length = document_lengths.max()
avg_length = document_lengths.mean()

# Print the results
print(f"Minimum document length: {min_length} words")
print(f"Maximum document length: {max_length} words")
print(f"Average document length: {avg_length:.0f} words")
```

```python
import matplotlib.pyplot as plt
```

```python

# Group by category
category_groups = df.groupby('Category')

# Initialize lists to store results
category_names = []
total_tokens_list = []
unique_tokens_list = []

# Loop through each category
for category, category_df in category_groups:
    documents = category_df['News'].tolist()

    # Tokenize and calculate counts
    vectorizer = CountVectorizer()
    X = vectorizer.fit_transform(documents)
    total_tokens = X.sum()
    unique_tokens = len(vectorizer.get_feature_names_out())

    # Store results
    category_names.append(category)
    total_tokens_list.append(total_tokens)
    unique_tokens_list.append(unique_tokens)

# Create a single bar graph for each category
plt.figure(figsize=(8, 8))

# Combine total tokens and unique tokens into a single bar
bar_width = 0.2
index = range(len(category_names))
plt.bar(index, total_tokens_list, width=bar_width, label='Total Tokens')
plt.bar([i + bar_width for i in index], unique_tokens_list, width=bar_width, label='Unique Tokens')

plt.xlabel('Category')
plt.ylabel('Counts')
plt.xticks([i + bar_width / 2 for i in index], category_names)
plt.legend()
plt.tight_layout()
plt.savefig('document_count_statistics.png')
plt.show()
```

```python

vectorizer = CountVectorizer()
category_stats = {'Category': [], 'Min Length': [], 'Max Length': [], 'Avg Length': []}

for category in df['Category'].unique():
    category_documents = df[df['Category'] == category]['News']

    X = vectorizer.fit_transform(category_documents)
    document_lengths = X.sum(axis=1)

    # Calculate statistics
    min_length = document_lengths.min()
    max_length = document_lengths.max()
    avg_length = int(document_lengths.mean())

    # Store statistics in the dictionary
    category_stats['Category'].append(category)
    category_stats['Min Length'].append(min_length)
    category_stats['Max Length'].append(max_length)
    category_stats['Avg Length'].append(avg_length)

# Create a DataFrame from the dictionary
category_stats_df = pd.DataFrame(category_stats)

# Plot a bar graph with three bars for each category
category_stats_df.plot(x='Category', kind='bar', figsize=(8, 10))
plt.xlabel('Category')
plt.ylabel('Document Length')
# plt.title('Category-wise Document Length Statistics')
plt.savefig('document_length_statistics.png')
plt.show()
```

```python
import pandas as pd
#load the dataset
df = pd.read_csv("/content/drive/MyDrive/Clean_BigCorpus.csv")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
documents = df['Document'].tolist()
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)
feature_names = vectorizer.get_feature_names_out()
total_tokens = X.sum()
unique_tokens = len(feature_names)

# Print the results
print(f"Total number of tokens (words): {total_tokens}")
print(f"Number of unique tokens (words): {unique_tokens}")
```

```python
#Average, min and max length
# business_documents = df[df['Category'] == 'Sports']['Document']
documents = df['Document'].tolist()
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)
feature_names = vectorizer.get_feature_names_out()
document_lengths = X.sum(axis=1)

# Calculate minimum, maximum, and average document lengths
min_length = document_lengths.min()
max_length = document_lengths.max()
avg_length = document_lengths.mean()

# Print the results
print(f"Minimum document length: {min_length} words")
print(f"Maximum document length: {max_length} words")
print(f"Average document length: {avg_length:.0f} words")
```

```python
import matplotlib.pyplot as plt
```

```python

# Group by category
category_groups = df.groupby('Category')

# Initialize lists to store results
category_names = []
total_tokens_list = []
unique_tokens_list = []

# Loop through each category
for category, category_df in category_groups:
    documents = category_df['Document'].tolist()

    # Tokenize and calculate counts
    vectorizer = CountVectorizer()
    X = vectorizer.fit_transform(documents)
    total_tokens = X.sum()
    unique_tokens = len(vectorizer.get_feature_names_out())

    # Store results
    category_names.append(category)
    total_tokens_list.append(total_tokens)
    unique_tokens_list.append(unique_tokens)

# Create a single bar graph for each category
plt.figure(figsize=(8, 8))

# Combine total tokens and unique tokens into a single bar
bar_width = 0.2
index = range(len(category_names))
plt.bar(index, total_tokens_list, width=bar_width, label='Total Tokens')
plt.bar([i + bar_width for i in index], unique_tokens_list, width=bar_width, label='Unique Tokens')

plt.xlabel('Category')
plt.ylabel('Counts')
plt.xticks([i + bar_width / 2 for i in index], category_names)
plt.legend()
plt.tight_layout()
plt.savefig('UDC_count_statistics.png')
plt.show()
```
